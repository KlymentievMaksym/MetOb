```python
import numpy as np
```

```python
def if_diag_pan(matrix, print_do=False):
    flag = True
    for row in range(matrix.shape[0]):
        flag &= abs(matrix[row, row]) >= sum(abs(matrix[row])) - abs(matrix[row, row])
    if print_do:
        print("Diag_Pan" if flag else "Not_Diag_Pan")
    return flag
```

```python
A = np.array([
    [6.59, 1.28, 0.79, 1.195, -0.21],
    [0.92, 3.83, 1.3, -1.63, 1.02],
    [1.15, -2.46, 5.77, 2.1, 1.483],
    [1.285, 0.16, 2.1, 5.77, -18],
    [0.69, -1.68, -1.217, 9, -6]
])
A_start = A
b = np.array([
    [2.1],
    [0.36],
    [3.89],
    [11.04],
    [-0.27]
])
b_start = b
A[1] += A[4]*1/4
b[1] += b[4]*1/4
# [ 0.25088613  0.79731537 -2.27511493  0.41375988]
A[2] += A[1]*0.8
b[2] += b[1]*0.8
# A[2] += A[1]*0.9565445
A[3] += A[4]*-2.28
b[3] += b[4]*-2.28
# A[3] += A[4]*-3.18926847
A[4] += A[3]*0.4
b[4] += b[3]*0.4
# A[4] += A[3]*0.20868201
if_diag_pan(A, print_do=True)
A, b
```

```python
C = np.zeros_like(A)
d = np.zeros_like(b)
for row in range(A.shape[0]):
    C[row] = (-A[row])/A[row, row]
    C[row, row] = 0
    d[row] = b[row]/A[row, row]
print(C)
print(d)
print()
print()
for row in range(C.shape[0]):
    print(sum(abs(C[row])), sum(abs(C[:, row])))
```

```python
def Iteration_simple_method(C, d, epsilon=1e-4):
    xk = np.zeros(b.shape)
    xk1 = C@xk + d

    count = 0

    r = abs(b_start - A_start@xk1)
    print(f"[{count}] r: \n {r} \n")
    print(f"[{count}] x: \n {xk1} \n\n")
    
    q = np.zeros(b.shape)
    for row in range(C.shape[0]):
        q = min(sum(abs(C[row])), sum(abs(C[:, row])))

    while q/(1-q) * np.max(abs(xk1 - xk)) >= epsilon:
        xk = xk1
        xk1 = C@xk + d
    
        count += 1
    
        r = abs(b_start - A_start@xk1)
        print(f"[{count}] r: \n {r} \n")
        print(f"[{count}] x: \n {xk1} \n\n")

    return xk1


x = Iteration_simple_method(C, d)

print(A_start@x)
print(b_start)
```
